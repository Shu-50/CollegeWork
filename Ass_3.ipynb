```python
import pandas as pd
df = pd.read_csv('Iris.csv')
df
```

```python
df.describe()
```

```python
group = df.groupby('Species')
group.describe()
```

```python
df['Species'] = df['Species'].str.strip().str.lower()
setosa = df[df['Species'] =='iris-setosa']
versicolor = df[df['Species'] == 'iris-versicolor']
virginica = df[df['Species'] == 'iris-virginica']
print("Statistics for Iris-setosa:")
print(setosa.describe())

print("\nStatistics for Iris-versicolor:")
print(versicolor.describe())

print("\nStatistics for Iris-virginica:")
print(virginica.describe())
```

```python
mean_value = df['PetalWidthCm'].mean()
mode_value = df['PetalWidthCm'].mode()
median_value = df['PetalWidthCm'].median()
std_dev = df['PetalWidthCm'].std()
variance = df['PetalWidthCm'].var()
range = df['PetalWidthCm'].max() - df['PetalWidthCm'].min()
```

```python
print('Mean: ',mean_value)
print('Mode: ',mode_value)
print('Median: ',median_value)
print('standard Deviation: ',std_dev)
print('Variance: ',variance)
print('Range: ',range)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Q1 = df['PetalWidthCm'].quantile(0.25)
Q2 = df['PetalWidthCm'].quantile(0.50)
Q3 = df['PetalWidthCm'].quantile(0.75)
IQR = Q3-Q1
print("\nQuartiles:")
print("Q1 :", Q1)
print("Q2 :", Q2)
print("Q3 :", Q3)
print("IQR :",IQR)
```

```python
plt.figure(figsize=(15, 10))
numerical_columns = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']
for i, column in enumerate(numerical_columns, 1):
    plt.subplot(2, 2, i)
    sns.histplot(df[column], kde=True, bins=15, color='skyblue')
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[numerical_columns], palette="Set2")
plt.title("Box Plot for Features")
plt.xlabel("Features")
plt.ylabel("Values")
plt.show()
```
