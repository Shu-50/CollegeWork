```python
import pandas as pd
df = pd.read_csv('Iris.csv')

df
```

```python
df.describe()
```

```python
group = df.groupby('Species')
group.describe()
```

```python
df['Species'] = df['Species'].str.strip().str.lower()
setosa = df[df['Species'] =='iris-setosa']
versicolor = df[df['Species'] == 'iris-versicolor']
virginica = df[df['Species'] == 'iris-virginica']
print("Statistics for Iris-setosa:")
print(setosa.describe())

print("\nStatistics for Iris-versicolor:")
print(versicolor.describe())

print("\nStatistics for Iris-virginica:")
print(virginica.describe())
```

```python
mean_value = df['PetalWidthCm'].mean()
mode_value = df['PetalWidthCm'].mode()
median_value = df['PetalWidthCm'].median()
std_dev = df['PetalWidthCm'].std()
```

```python
print('Mean: ',mean_value)
print('Mode: ',mode_value)
print('Median: ',median_value)
print('standard Deviation: ',std_dev)
```
